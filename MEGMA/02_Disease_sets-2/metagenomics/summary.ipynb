```python
import glob
import pandas as pd
import os
```

```python
data_list = glob.glob('./*.csv')
```

```python
ts = pd.Series(data_list).sort_values().reset_index(drop=True)
```

```python
df = ts.to_frame(name = 'path')
```

```python
df['data_name'] = df.path.apply(lambda x:os.path.basename(x).split('_')[0])
```

```python
df.data_name.unique()
```

```python
data_name_map = {'cirphy':'Cirrhosis_Nguyen', 
 'colphy':'COL_Nguyen', 
 'ibdphy':'IBD_Nguyen', 
 'obephy':'Obesity_Nguyen', 
 't2dphy':'T2D_Nguyen', 
 'wt2dphy':'WT2D_Nguyen'}
```

```python
df['Datasets'] = df.data_name.map(data_name_map)
```

```python
res = []
for dataset in df.Datasets:
    df1 = df[df.Datasets == dataset]
    dfx = pd.read_csv(df1.path.iloc[0], index_col=0)
    dfy = pd.read_csv(df1.path.iloc[1], index_col=0)
    
    P, N = dfx.shape
    
    des = dfy.x.map({0:'Control', 1:'Case'}).value_counts().to_dict()
    des['Species'] = P
    des['Dataset'] = dataset
    res.append(des)
```

```python
pd.DataFrame(res).drop_duplicates().set_index('Dataset')
```
